### 1. Carregar o dataset tratado

```python
import pandas as pd

# Carregar o dataset limpo para an√°lise descritiva
try:
    dados_abc = pd.read_excel("dataset_tratadoABC.xlsx")
    print("Dataset tratado carregado para An√°lise ABC.")
except FileNotFoundError:
    print("ERRO: Certifique-se de que o arquivo 'dataset_tratado.xlsx' est√° na mesma pasta.")
    exit()

    print("-" * 50)
print("MINERA√á√ÉO 4.1: AN√ÅLISE ABC (PARETO) POR CATEGORIA")
print("-" * 50)
```

### 1. Agrega√ß√£o: Calcular a Receita Total por Categoria

```python
vendas_por_categoria = dados_abc.groupby('Categoria')['Total de Vendas'].sum().reset_index()
vendas_por_categoria.columns = ['Categoria', 'Receita_Total']
```

### 2. Ordenar e Calcular o Total Geral

```python
vendas_por_categoria = vendas_por_categoria.sort_values(by='Receita_Total', ascending=False)
total_receita_geral = vendas_por_categoria['Receita_Total'].sum()
```

### 3. Calcular Porcentagem de Contribui√ß√£o e Acumulada

```python
vendas_por_categoria['%_Contrib_Receita'] = (vendas_por_categoria['Receita_Total'] / total_receita_geral) * 100
vendas_por_categoria['Receita_Acumulada'] = vendas_por_categoria['Receita_Total'].cumsum()
vendas_por_categoria['%_Acumulada'] = (vendas_por_categoria['Receita_Acumulada'] / total_receita_geral) * 100
```

### 4. Exibir o Resultado (Curva ABC)

```python

print("An√°lise ABC (Pareto) - Ranqueamento de Receita")
print(vendas_por_categoria.round(2).to_markdown(index=False))
```

### 5. Classifica√ß√£o (Insight)

```python
categoria_A = vendas_por_categoria[vendas_por_categoria['%_Acumulada'] <= 80]
categoria_B = vendas_por_categoria[(vendas_por_categoria['%_Acumulada'] > 80) & 
                                   (vendas_por_categoria['%_Acumulada'] <= 95)]
categoria_C = vendas_por_categoria[vendas_por_categoria['%_Acumulada'] > 95]

print("\n--- Resumo da Classifica√ß√£o ABC ---")
print(f"üèÜ Categoria A (Top 80% Receita): {', '.join(categoria_A['Categoria'])}")
print(f"üí∞ Categoria B (Pr√≥ximos 15% Receita): {', '.join(categoria_B['Categoria'])}")
print(f"üìâ Categoria C (√öltimos 5% Receita): {', '.join(categoria_C['Categoria'])}")

print(f"\nüí° INSIGHT: {len(categoria_A)} categorias s√£o respons√°veis por 80% do faturamento da loja, exigindo prioridade na gest√£o de estoque.")
```
